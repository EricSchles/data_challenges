The first thing I did was describe the data numerically:


```python
import pandas as pd
import numpy as np
from scipy import stats

def describe_numerical_column(column_data):
    print("Mean", column_data.mean())
    print("Std Dev", column_data.std())
    print("Skew", column_data.skew())
    print("Kurtosis", column_data.kurtosis())
    print("Normal Test Results",stats.normaltest(column_data))


def describe_data(df):
    for column in df.columns:
        if df[column].dtype == np.float64 or df[column].dtype == np.int64:
            print(column)
            describe_numerical_column(df[column])
        else:
            print(column)
            print(df[column].value_counts())

describe_data(df)
```

```python
import numpy as np

def numeric_only_data(df):
    numeric_columns = []
    for column in df.columns:
        if df[column].dtype == np.float64 or df[column].dtype == np.int64:
            numeric_columns.append(column)
    return numeric_columns

tmp_df = df[df["visa_class"] == "H-1B"]
tmp_df.dropna(inplace=True)
categorical_columns = ["status", "lca_case_employer_state", "lca_case_soc_name", "lca_case_wage_rate_unit", "full_time_pos"]
numeric_columns = numeric_only_data(df)
columns = df.columns.tolist()
columns = [column for column in columns if column in numeric_columns]
categorical_df = pd.get_dummies(tmp_df[categorical_columns])
numeric_df = tmp_df[numeric_columns]
```

## Digging into the dependent variable


For this example the dependent variable is the `lca_case_wage_rate_from`. I used pearson correlation to look at how wages were correlated with variables in the dataset.  

```python
from scipy import stats
import time


def corr_categorical_with_wages(df, wages):
    correlations = {}
    columns = df.columns.tolist()
    start = time.time()
    for col in columns:
        correlations["wages" + '___' + col] = stats.pointbiserialr(wages, df[col].values)
    results = pd.DataFrame.from_dict(correlations, orient="index")
    results.columns = ["correlation", "pvalues"]
    results.sort_index(inplace=True)
    return results

def corr_numeric_with_wages(df, wages):
    correlations = {}
    columns = df.columns.tolist()
    start = time.time()
    for col in columns:
        correlations["wages" + '___' + col] = stats.spearmanr(wages, df[col].values)
    results = pd.DataFrame.from_dict(correlations, orient="index")
    results.columns = ["correlation", "pvalues"]
    results.sort_index(inplace=True)
    return results

cat_corr_df = corr_categorical_with_wages(categorical_df, numeric_df["pw_1"].values)
num_corr_df = corr_categorical_with_wages(numeric_df, numeric_df["pw_1"].values)
print("categorical data")
for i in cat_corr_df.index:
    if cat_corr_df.ix[i]["pvalues"] < 0.05:
        print(i, cat_corr_df.ix[i]["correlation"])
print("numeric data")
for i in num_corr_df.index:
    if num_corr_df.ix[i]["pvalues"] < 0.05:
        print(i, num_corr_df.ix[i]["correlation"])
```

As we can see you can, see there not all the variables above have statistically significant correlations with wages.  So, we segment out those with p-values above 0.06.

Observing the above we can see the variables of most interest and strongest correlation with wage.  We can run a decision tree to see how these different variables contribute to the decision process of wage setting.  

Before we dive into modeling though - we can do two things - remove some variables from our model, namely those variables that had a p-value above 0.05.  And we can perform PCA to reduce the number of dummy variables that we need to account for.

```python
from sklearn import decomposition
import matplotlib.pyplot as plt

variables_to_remove = []
def process_index(i):
    return i.replace("wages___","")

for i in cat_corr_df.index:
    if cat_corr_df.ix[i]["pvalues"] > 0.05:
        variables_to_remove.append(process_index(i))

categorical_df = categorical_df[variables_to_remove]
#typically we need to standardize but these are all dummy variables
pca = decomposition.PCA().fit(categorical_df)

def screeplot(pca, standardised_values):
    y = np.std(pca.transform(standardised_values), axis=0)**2
    x = np.arange(len(y)) + 1
    y = y[:10]
    x = x[:10]
    plt.plot(x, y, "o-")
    plt.xticks(x, ["Comp."+str(i) for i in x], rotation=60)
    plt.ylabel("Variance")
    plt.show()

screeplot(pca, categorical_df)
```

There are form more components than this, but with PCA we follow the "elbow" rule, when the graph starts to level off, that's all the components you need!  In this case it looks like 2 or 3 will do the trick.  We'll take 3 components to be safe.

```python
transformed_categories = pca.transform(categorical_df)
pc1 = transformed_categories[:, 0]
pc2 = transformed_categories[:, 1]
pc3 = transformed_categories[:, 2]
categorical_df = pd.DataFrame({"PC1":pc1, "PC2":pc2, "PC3":pc3})
categorical_df
```

```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras import regularizers

from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import math
from sklearn import metrics
from functools import partial
from keras import optimizers

# fix random seed for reproducibility
seed = 7
np.random.seed(seed)

def feature_transform(df, col_name):
    df[col_name+"_squared"] = df[col_name].apply(lambda x: x*x)
    df[col_name+"_log"] = df[col_name].apply(lambda x: math.log(x) if x != 0 else x)
    df[col_name+"_sin"] = df[col_name].apply(lambda x: math.sin(x))
    df[col_name+"_sqrt"] = df[col_name].apply(lambda x: math.sqrt(x))
    try:
        df[col_name+"_gamma"] = df[col_name].apply(lambda x: math.gamma(x) if x!=0 else x)
    except OverflowError:
        pass
    return df

# larger model
def create_network():
    # create model
    reg_rate = 0.01
    model = Sequential()
    model.add(Dense(250, 
                    input_dim=117, 
                    activation="linear",
                    #kernel_regularizer=regularizers.l2(reg_rate),
                    activity_regularizer=regularizers.l1(reg_rate)))
    model.add(Dense(500, 
                    activation="linear",
                    #kernel_regularizer=regularizers.l2(reg_rate),
                    activity_regularizer=regularizers.l1(reg_rate)))
    model.add(Dense(500,  
                    activation="linear",
                    #kernel_regularizer=regularizers.l2(reg_rate),
                    activity_regularizer=regularizers.l1(reg_rate)))
    model.add(Dense(500,  
                    activation="linear",
                    #kernel_regularizer=regularizers.l2(reg_rate),
                    activity_regularizer=regularizers.l1(reg_rate)))
    model.add(Dense(1, activation="linear"))
    # Compile model
    adam = optimizers.Adam(lr=0.01, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
    model.compile(loss='mean_squared_error', optimizer=adam)
    return model

X_train_for_nn = X_train.as_matrix()
y_train_for_nn = y_train.as_matrix()
X_test_for_nn = X_test.as_matrix()
y_test_for_nn = y_test.as_matrix()

estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=create_network, epochs=25, batch_size=30, verbose=0)))
pipeline = Pipeline(estimators)
#kfold = StratifiedKFold(n_splits=10, random_state=seed)
#results = cross_val_score(pipeline, X_train_for_nn, y_train_for_nn, cv=kfold)
#print("Neural Network: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
model = pipeline.fit(X_train_for_nn, y_train_for_nn)
#print("Results of cross validation:")
#print(results)
#print("Mean score of cross validation:")
#print(np.mean(results))
#print("For activation function", activation)
print("Model Score:")
print(model.score(X_test_for_nn, y_test_for_nn))
#predictions = model.predict(X_test_for_nn)
#print("RMSE Results", math.sqrt(metrics.mean_squared_error(y_test.values, predictions)))
```
